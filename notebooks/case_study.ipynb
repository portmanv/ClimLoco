# Description
This notebook uses CLimLoco1.0 to a case study on the global mean surface air temperature (GSAT), to produce the figures and values of the corresponding appendix in the article.
- Imports the simulated data (from CMIP6), the observed data (from HadCRUT5).
- Defines Y as the mean 2081-2100 GSAT, relative to 1850-1900.
- Searches different X correlated to Y. Finds the 1970-2014 GSAT mean (X1) and the 1970-2014 GSAT trend (X2).
- Compares ClimLoco1.0 results when using X1, X2, or both in the same time.
- Display the distributions of Y, X1 and X2, to evaluate if they are gaussian.
- Compares ClimLoco1.0 results with the results using two other methods of the literature, in different configurations of the data.
- Displays the figure of the annex explaining the key statistical concepts.

# Comparison with two existing methods
### Methods "Bayesian" (Bowman et al. (2018), Ribes et al. (2021))
These methods lead to the same interval as our, with the difference to neglecting the fact that there is a limited number of climate models M.

### Method linear from Cox et al. (2018)
Cox et al. obtain an interval of : $\hat{a}_0 + \hat{a}_1 \,x_0 \pm z\, \sqrt{\hat{\sigma}_{\varepsilon}^2 \, (1+\frac{1}{M} + \frac{(x_0 - \hat{\mu}_X)^2}{M\,\hat{\sigma}_X^2}) + a_1^2 \, \sigma_N^2}$

Where ClimLoco1.0 gives the interval: $\hat{b}_0 + \hat{b}_1 \,x_0 \pm t^{M-2} \, \hat{\sigma}_{\varepsilon^N} \, \sqrt{1+\frac{1}{M} + \frac{(x_0 - \hat{\mu}_X)^2}{M\,(\hat{\sigma}_X^2+\sigma_X^2)}}$

z and t are gaussian and student quantiles. A major difference is that Cox et al. 2018 considers the regression of $X$ on $Y$, while we consider the regression of $X^N$ (observational noise) on $Y$:
- $Y = a_0 + a_1\,X + \varepsilon$
- $Y = b_0 + b_1\,X^N + \varepsilon^N$


In the example of $X_1$ that constraint Y, we obtain:

```python
print("Our result:")
[_,_] = compute_confidence_intervals_X_multivariate(X1, X1_obs, Y, 0.90, period_Y, reference_period, display=True)
```

```python
print("Result neglecting the limited sample size (Bowman et al. (2018)):")
[_,_] = compute_confidence_intervals_X_multivariate(X1, X1_obs, Y, 0.90, period_Y, reference_period, neglect_M=True, display=True)
```

```python
print("Result obtained using the method from Cox et al. 2018:")

var_N = np.var(X1_obs, ddof=1)
constrained_CI = compute_Cox_confidence_intervals(X1, np.mean(X1_obs).reshape(-1,1), Y, 0.9, var_N)

print("Constrained : [{:.2f} +- {:.2f}]".format(np.mean(constrained_CI), np.diff(constrained_CI)[0]/2))
```

# Test if our implementation of Cox et al. (2018) is correct

By using the data from the article Cox et al. (2018), to find the same interval [2.2, 3.4] (66% confidence) of the ECS constrained by Ψ.

```python
# Data from Extended Data Table 1 of Cox et al. (2018)
# Columns : λ (Wm⁻² K⁻¹), ECS (K), Q₂ₓCO₂/σₙ, Ψ (K)
# Rows : climate models
data = np.array([
    [0.8, 3.8, 8.5, 0.22],
    [1.0, 3.7, 8.3, 0.17],
    [1.2, 2.9, 7.3, 0.19],
    [1.1, 3.3, 8.7, 0.16],
    [0.6, 4.1, 6.1, 0.21],
    [1.4, 2.4, 5.9, 0.15],
    [0.6, 4.6, 7.8, 0.29],
    [1.4, 2.1, 11.9, 0.07],
    [1.0, 2.6, 7.2, 0.16],
    [0.9, 4.7, 11.7, 0.23],
    [1.1, 3.6, 11.9, 0.15],
    [1.2, 2.6, 9.3, 0.09],
    [1.1, 2.8, 7.8, 0.14],
    [1.1, 2.8, 6.9, 0.19],
    [1.8, 2.1, 11.1, 0.12],
    [1.0, 4.1, 8.0, 0.15],
    [1.3, 2.4, 7.1, 0.20],
    [0.8, 4.0, 6.7, 0.36],
    [0.8, 4.1, 8.6, 0.20],
    [1.5, 2.7, 10.2, 0.23],
    [1.2, 2.9, 7.4, 0.14],
    [1.7, 2.3, 11.8, 0.10]
])

# only the 16  first models are used in their article (see their figure 2.b in Cox et al. (2018))
ECS = data[:16, 1]
phi = data[:16, 3]

# According to the Extended Data Table 2 in Cox et al. (2018), HadCrut4 gives an observed value of Ψ of mean 0.13 and standard deviation of 0.016.
phi_obs = np.array(0.13)
var_phi_obs = np.square(0.016)
compute_Cox_confidence_intervals(phi.reshape(-1,1), phi_obs.reshape(-1,1), ECS, 0.66, var_phi_obs)
```

We correctly obtained [2.2, 3.4] (with a 66% confidence) !
